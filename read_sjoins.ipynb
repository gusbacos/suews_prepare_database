```python
import os
import glob
import pandas as pd
path = 'C:/GitHub/SUEWS/docs/source/input_files/'
csvs = path + 'SUEWS_SiteInfo/csv-table/'

```

```python
dict = {}

for file in glob.glob("*.csv"):
    df_csv = pd.read_csv(csvs + file)
    param = file[:-4]
    dict[param.lower()] = {}
    try:
        if len(df_csv)> 1:
            dict[param.lower()] = {
                'SUEWS' : param,
                'Referencing Table' : {}
                }

            for i in range(len(df_csv)):
                dict[param.lower()]['Referencing Table'][df_csv.iloc[i]['Referencing Table']] = {'Comment' : df_csv.iloc[i]['Comment']}   
        else:
            dict[param.lower()] = {
                'SUEWS' : param,
                'Referencing Table' : {
                    df_csv['Referencing Table'].item() : {
                        'Comment' : df_csv['Comment'].item()
                        }
                    }   
                }   

    except:
        print(param)
```
